# 3. **Prepare the data.**

* Take care of missing data (if you have not done this already)
* Data cleaning (remove outliers, if necessary).
* Feature engineering (create new features/ delete non relevant ones/ etc. Iterative process)


```python
# Since there are no missing values, we can proceed
# Check for outliers using z-score
from scipy import stats
import numpy as np

z_scores = np.abs(stats.zscore(diabetes.select_dtypes(include=[np.number])))
outliers = (z_scores > 3).any(axis=1)
print("Number of outliers:", outliers.sum())

# Remove outliers
diabetes_clean = diabetes[~outliers]

# Feature engineering (if necessary)
# For this dataset, we can proceed without creating new features
```

# 4. Divide data into ``train/ test`` or ``train/validation/test`` datasets

* Transformation (scale your data, if needed). 

```python
from sklearn.model_selection import train_test_split

# Define features and target variable
X = diabetes_clean.drop('Diabetes_binary', axis=1)
y = diabetes_clean['Diabetes_binary']

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Transformation (scale your data, if needed)
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# 5. Models building and comparison

* Select the correct evaluation metric or metrics (https://scikit-learn.org/stable/modules/model_evaluation.html)
* Remember to always use the same random_state to be able to repeat results.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

# Initialize models with random_state=42
models = {
    'Logistic Regression': LogisticRegression(random_state=42),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(random_state=42)
}

# Train models and evaluate
for name, model in models.items():
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)
    print(f"Results for {name}:")
    print(classification_report(y_test, y_pred))
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    if hasattr(model, "predict_proba"):
        y_proba = model.predict_proba(X_test_scaled)[:,1]
        print("ROC AUC Score:", roc_auc_score(y_test, y_proba))
    print("\n")
```

* Which model(s) obtains better result(s)?


```python
#Based on the results:

#Logistic Regression achieves the highest ROC AUC score of 0.818. It has high precision and recall for class 0 (no diabetes) but low recall (0.11) for class 1 (diabetes), indicating difficulty in identifying diabetic cases.
#Decision Tree has a lower ROC AUC score of 0.588 but a slightly better recall for class 1 (0.28) compared to the other models. However, its overall performance is inferior.
#Random Forest attains a ROC AUC score of 0.778, with high performance for class 0 but, like Logistic Regression, low recall for class 1.

#The Logistic Regression model obtains the better overall results based on the ROC AUC score.
```

# 6. Adjust hyperparameters.

Steps to perform the hyperadjustment of the parameters in the best model or models selected from step 5):

* Select the Evaluation Metrics
* Define search hyperparameter ranges: model params


```python
from sklearn.model_selection import GridSearchCV

# For the best model (suppose Random Forest performed the best)
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'max_features': ['auto', 'sqrt']
}

rf = RandomForestClassifier(random_state=42)
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, scoring='roc_auc', n_jobs=-1)

grid_search.fit(X_train_scaled, y_train)
print("Best parameters found:", grid_search.best_params_)
```

# **7. Final evaluation of the model**

Train the model with the optimal hyperparameters found in the previous section and make the predictions.



```python
best_rf = grid_search.best_estimator_
best_rf.fit(X_train_scaled, y_train)

# Make predictions
y_pred = best_rf.predict(X_test_scaled)
y_proba = best_rf.predict_proba(X_test_scaled)[:,1]

# Evaluate the model
print("Classification Report:")
print(classification_report(y_test, y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("ROC AUC Score:", roc_auc_score(y_test, y_proba))
```

#### Model results conclusions

```python
#After hyperparameter tuning and retraining:

#The model maintains a high precision and recall for class 0 but has a very low recall (0.04) for class 1.
#The ROC AUC score remains at 0.818, indicating no significant improvement in distinguishing between classes.

#The model is still not effectively identifying diabetic cases. Despite adjustments, it is biased towards the majority class.
```
